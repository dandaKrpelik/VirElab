```python
import numpy as np
import matplotlib.pyplot as plt
import osci as osci
```

```python
o = osci.Oscilloscope()

s1 = osci.Signal(spectrum = [[50, 3]], noise = 0.05)
s2 = osci.Signal(square = [[125, 2, 0.7], [20, 2, 0.2]])

c1 = osci.Channel(s1)
c1.voltdiv = 3 * 2/ o.vdiv
c2 = osci.Channel(s2)
c2.voltdiv = 1

o.add_channel(c2)
o.add_channel(c1)


o.step()

c1.dh = 0.01
o.show()
```

```python
o = osci.Oscilloscope()

s1 = osci.Signal(spectrum = [[50, 2.5]], noise = 0.05)
s2 = osci.Signal(spectrum = [[200, 1.5]], noise = 0.05)

c1 = osci.Channel(s1)
c1.voltdiv = 0.5
c2 = osci.Channel(s2)
c2.voltdiv = 0.5

o.add_channel(c1)
o.add_channel(c2)

o.mode = [c1,c2]

o.step()
o.show()
```

```python
s2 = osci.Signal(spectrum = [[150, 1.5]], noise = 0.01)
c2.signal = s2

o.step()
o.show()
```
